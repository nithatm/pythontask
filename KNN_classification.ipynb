# importing libraries

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
```

# loading data

```python
data=pd.read_csv('/content/emails.csv')
data.head()
```

```python
data.describe()
```

```python
data[data.isnull().any(axis=1)]
```

 to separate columns in a dataset into categorical and numerical variables based on their data types.

```python
categorical=[var for var in data.columns if data[var].dtype=='O']
numerical=[var for var in data.columns if data[var].dtype!='O']
print(f'there are {len(categorical)} variables :{categorical}')
```

check the cardinality of categorical variable

```python
for var in categorical:
  print(var,'contains',len(data[var].unique()),'labels')
```

statistical summary of numerical variables

```python
print(round(data[numerical].describe()),2)
```

data processing

```python
from sklearn.preprocessing import LabelEncoder
data['Email No.']=LabelEncoder().fit_transform(data['Email No.'])
data.head()
```

assigning values to x &Y

```python
y=data['Prediction']
x=data.drop(['Prediction'],axis=1)
```

```python
print(x.head())
```

```python
print(y.head())
```

splitting the data into training and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
print(x_train.head())
```

```python
print(x_test.head())
```

```python
print(y_train[:10])
```

```python
print(y_test[:10])
```

 for standardizing features by removing the mean and scaling to unit variance.Standardization of datasets is a common requirement for many machine learning algorithms. It helps to center the data around 0 and scale it to have a standard deviation of 1.

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

# building a model

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=2,p=2,metric='minkowski')# by default n_neighbors = 5
classifier.fit(x_train,y_train)
```

# prediction

```python
y_pred=classifier.predict(x_test)
```

# evaluate model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,average_precision_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
cr = classification_report(y_test,y_pred)
print(cr)
```

```python
average_precision= average_precision_score(y_test,y_pred)
print(average_precision)
```
